## Mode- Imputation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train2.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
df.head()
```

```python
df.sample(10)
```

## Checking Missing (NULL) Value

```python
df.isnull().mean()*100
```

## Plot Bar Garage Value

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
df['GarageQual'].mode()
```

## KDE Plot | Compare Housees with TA | Null

```python
fig=plt.figure()
ax=fig.add_subplot(111)
df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax)
df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,color='red')
lines, lables=ax.get_legend_handles_labels()
labels=['Houses With TA','Houses With NA']
ax.legend(lines,lables,loc='best')
plt.title('GarageQual')
```

## Store Variable TA in temp

```python
temp = df[df['GarageQual'] == 'TA']['SalePrice']
```

## Replace Missing value with TA

```python
df['GarageQual'].fillna('TA',inplace=True)
```

## Review Bar Plot changes

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

## Draw plot again | After imputation

```python
fig=plt.figure()
ax=fig.add_subplot(111)
temp.plot(kind='kde',ax=ax)

#distribution of the variable afteer imputation

df[df['GarageQual'] == 'TA']['SalePrice'].plot(kind='kde',ax=ax,color='red')
lines, labels=ax.get_legend_handles_labels()
labels= ['Original Variable','Imputed Variable']
ax.legend(lines,labels,loc='best')

#adding title
plt.title('GarageQual')
```

## Plot bar fire place

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
```

```python
df['FireplaceQu'].mode()
```

## KDE Plot | Replace House with GD & NA

```python
fig=plt.figure()
ax=fig.add_subplot(111)
df[df['FireplaceQu'] == 'Gd']['SalePrice'].plot(kind='kde',ax=ax)
df[df['FireplaceQu'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,color='red')
lines, labels=ax.get_legend_handles_labels()
labels=['Houses with Gd','Houses with NA']
ax.legend(lines,labels,loc='best')
plt.title('FireplaceQu')
```

## Store Temp Variable

```python
temp=df[df['FireplaceQu']=='Gd']['SalePrice']
```

## Replace missing value with GD

```python
df['FireplaceQu'].fillna('Gd',inplace=True)
```

## Draw Bar Plot

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
```

## Draw plot again | After imputation

```python
fig=plt.figure()
ax=fig.add_subplot(111)
temp.plot(kind='kde',ax=ax)

#Distribution of the variable after imputation

df[df['FireplaceQu']=='Gd']['SalePrice'].plot(kind='kde',ax=ax,color='red')
lines,labels=ax.get_legend_handles_labels()
labels=['Original Variable','Imputed Variable']
ax.legend(lines,labels,loc='best')

#add title
plt.title('FireplaceQu')
```
